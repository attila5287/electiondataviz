```python
import random
import time


def print_pause(message_to_print):
    print(message_to_print)
    time.sleep(2)


def valid_input(prompt, option1, option2):
    while True:
        response = input(prompt).lower()
        if option1 in response:
            break
        elif option2 in response:
            break
        else:
            print_pause("sorry, I don't understand.")
        return response


def intro():
    print_pause("You find yourself standing in an open field, "
                "filled with grass and yellow wildflowers.")
    print_pause("Rumor has it that a wicked fairie is somewhere around here, "
                "and has been terrifying the nearby village...")
    print_pause("In front of you a house."
                "To your right is a dark cave.")
    print_pause("In your hand you hold your trusty "
                "(but not very effective) dagger.\n")


def cave(items):
    print_pause("You peer cautiously into the cave. "
                "It turns out to be only a very small cave. "
                "Your eye catches a glint of metal behind a rock.")
    print_pause("You have found the sword of dead king! ")
    print_pause("You discard your silly old dagger, "
                "and take the sword with you. "
                "You walk back out to the field.")
    if "sword" in items:
        print_pause("Already, You've got the sword, "
                    "there is nothing to do here!")
    else:
        print_pause("You have found the sword of dead king!"
                    "You discard your silly old dagger, "
                    "and take the sword with you. "
                    "You walk back out to the field.")
        items.append("sword")
    field(items)


def house(items):
    print_pause("You approach the door of the house.")
    print_pause("You are about to knock when the door "
                "opens and out steps a gorgon. "
                "Oops! This is the gorgon's house! The gorgon attacks you!")

    fight(items)


def fight(items):
    print_pause("Would you like to (1) fight or (2) run away?")
    fight = input(
        "1\n"
        "2\n")
    if fight == '1':
        if "tunnel" in items:
            print_pause("You feel a bit under-prepared for this, "
                        "what with only having a tiny dagger.")
            print_pause("You do your best...")
            print_pause("But your dagger is no match for the gorgon.\n"
                        "You have been DEFEATED!\n")
        else:
            if "sword" in items:
                print_pause(
                    "As the pirate moves to attack, "
                    "you unsheath your new sword.\n"
                    "The sword of Ogoroth shines "
                    "brigthly in your hand \n"
                    "as you brace yourself for the attack.\n"
                    "But the pirate takes one look at your "
                    "shiny new toy and runs away!\n"
                    "You have rid the town of pirate.\n")
                print_pause("You're VICTORIOUS!\n")
                items.append("tunnel")
            else:
                print_pause("You feel a bit under-prepared for this, "
                            "what with only having a tiny dagger.")
                print_pause("You do your best...")
                print_pause("But your dagger is no match for the gorgon.\n"
                            "You have been DEFEATED!\n")
                play_again()

    if fight == '2':
        print_pause("You run back into the field. Luckily, you don't seem"
                    "to have been followed.")
        field(items)


def field(items):
    print_pause("Enter 1 to knock on the door of the house.\n"
                "Enter 2 to peer into the cave.\n"
                "What would you like to do?\n"
                "(Please enter 1 or 2). \n")
    enter = input(
        "1\n"
        "2\n")
    if enter == '1':
        house(items)
    elif enter == '2':
        cave(items)


def play_game():
    items = []
    intro()
    field(items)


def play_again():
    response = input(
        "Would you like to play again?\n"
        "Please Enter \"yes\" or \"no\".\n")
    if response == "yes":
        play_game()
    if response == "no":
        exit()


play_game()

play_again()
```
